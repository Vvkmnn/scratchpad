```python
A = [ 20, 1, -1, 2, -2, 3, 3, 5, 5, 1, 2, 4, 20, 4, -1, -2, 5 ]
n =  5

def find_it(seq):
    """
    Given a sequence of numbers, find the original binary representation of the first number, then compare it with
    all the other numbers one by one using bitwise xor, which will return 1 if the bits are the same and 0 if the bits
    are different. Therefore, to find the binary sequence that appears most often, we find the bits that return 1,
    which will be leftover after every xor flip.
    """
    res = 0
    for element in seq:
        print("Comparing {} to {}".format(bin(res),bin(element)))
        res = res ^ element
        print("New res is {}".format(bin(res)))
    return res

find_it(A)
```

### Give me Diamond

```python
def diamond(n):
    """
    Smrater code that does the same loop for line generation, skips
    checking for strings, and uses n%2 == 1 to test for odd numbers.
    
    The string uses half the length for spaces, but apppends the string 
    by accounting for spaces as well. Only seems to work for Python 2. 
    
    """
    if n > 0 and n % 2 == 1:
        diamond = ""
        for i in range(n):
            diamond += " " * abs((n/2) - i)
            diamond += "*" * (n - abs((n-1) - 2 * i))
            diamond += "\n"
        return diamond
    else:
        return None
    
diamond(3)
```

```python
def prime_factors(n):
    i = 2
    factors = []
    while i * i <= n:
        print('While factor {} squared ({}) is less than {}, check the following:'.format(i, i ** 2, n))
        if n % i: 
            print('{} mod {} is not 0; increase the factor by 1 until we get a factor that is mod 0'.format(n, i, n % i))
            i += 1
        else:
            print('{} mod {} is 0, so we floor divide {} by {}, yielding new n {}'.format(n, i, n // i))
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors

prime_factors(9)
```

### [Binary to Text (ASCII) Conversion](https://www.codewars.com/kata/5583d268479559400d000064/train/python)



```python
def binary_to_string(binary):
    """
    Faster character conversion method, which converts every 8 step set of bits until the
    end of the binary string into an integer of base 2, and then converts that into its
    ASCII character. 
    """
    ## Function Exploration: 
    
    # Chunk every 8 bit section of the binary function, up to the end 
    print([ binary[i: i+8] for i in range(0, len(binary), 8)])
    
    # Convert that into an integer of base 2, to get the right ASCII character reference (Too many bits to convert otherwise)
    print([ int(binary[i: i+8], 2) for i in range(0, len(binary), 8)])
    
    # Convert those references into a list of characters
    print([ chr(int(binary[i: i+8], 2)) for i in range(0, len(binary), 8)])
    
    # Finally, join those characters to an empty string and return that string 
    print("".join( [ chr( int(binary[i: i+8], 2) ) for i in range(0, len(binary), 8) ] ))
    
    ## Final Answer: 
    return "".join( [ chr( int(binary[i: i+8], 2) ) for i in range(0, len(binary), 8) ] )
    
binary_to_string('0100100001100101011011000110110001101111')
```

### [Unique In Order](https://www.codewars.com/kata/54e6533c92449cc251001667)

```python
def unique_in_order(iterable):
    """
    A best practice solution. Same previous to current comparsion, but m
    """
    result = []
    prev = None
    for char in iterable[0:]:
        if char != prev:
            result.append(char)
            prev = char
    return result
```

### [Tribonacci Sequence](https://www.codewars.com/kata/556deca17c58da83c00002db/train/python)



```python
def tribonacci(signature, n):
    """
    Same idea, but with some fancier list syntax early on to avoid problems:
    * Creates a result list of length n `signature[:n]`, which will include 3 if n > 3, or less if n < 2, avoiding the 
    `result[0:n]` problem. 
    * Single line for loop, using default 0 start and range(n-3) instead of range (3,n)
    * Same append logic, but with `sum[-3:]`, which PICKS THE LAST 3 VARIABLES IN A LIST and sums them. Much
    handier than `result[(x-3):x]`, which is the same logic. 
    """
    res = signature[:n]
    for i in range(n - 3): res.append(sum(res[-3:])) #; print(res[-3:]) Using ; to end line and run print command
    return res
```

```python
def arrange(strng):
    """
    * Split the string into a list seperated by spaces
    * For every odd try starting at 0, `i%2 != 0` where current word is shorter than next, switch 'em via `l.insert(newindex, l.pop(oldindex))`
    * For every even try, with the word longer than the next, switch them the same way
    * Join up the final list with spaces to get a string, alternating upper and lower case for each word
    * Note that this is not a list comprehension, where `if` would be at the end, this is a ternary operation (do a if true, else b)
    * Ternary operation in list mode: [doSomething() if (condition) else doSomethingElse() for value in list]
    """
    source = strng.split(' ')
    for i in range(len(source)-1):
        if i%2 != 0  and (len(source[i]) < len(source[i+1])):
            source.insert(i+1, source.pop(i))
        if i%2 == 0  and (len(source[i]) > len(source[i+1])):
            source.insert(i+1, source.pop(i))
    return(" ".join([source[x].upper() if x%2 != 0 else source[x].lower() for x in range(len(source))]))
```

### [Tic-Tac-Toe Checker](https://www.codewars.com/kata/525caa5c1bf619d28c000335/train/python)

```python
def isSolved(board):
    """
    An interesting solution thats trades raw readability for a more optimized return, 
    skipping the last `if` block in my code below, and returning the required values
    without any required transformations. Very clever setup that shows the power of using
    array indices: `array[rowindex][colindex]`.
    
    * First loop (`board[i][0]) checks if each element in the current row is the same and not 0; if it is, returns the first value
    * Second loop (`board[0][i]`) checks by column (very clever), and sees if the values are the same and not 0; if so, return that value
    * Once its looped through, it jumps to an if block that checks diagonals using the same logic (diag indices the same and not 0? altdiag the same and not 0?)
    * Finally, it uses the `0 not in [list]` and checks every row by primary index `board[i]` returns 0
    """
    # Loop by 3 to get in 3 rows and 3 columns
    for i in range(0,3):
        # Check if elements are the same by row indicies [0,1,2] and not 0, and return the first element 
        if board[i][0] == board[i][1] == board[i][2] != 0:
            return board[i][0]
        # Check if elements are the same by column indicies [0,1,2] and not 0, and return  first element 
        elif board[0][i] == board[1][i] == board[2][i] != 0:
            return board[0][i]
     
    # Check if elements are the same by diagonal indicies and not 0, and return the  first element 
    if board[0][0] == board[1][1] == board[2][2] != 0:
        return board[0][0]
    # Check if elements are the same by the alternative diagonal indicies and not 0, and return the  first element 
    elif board[0][2] == board[1][1] == board[2][0] != 0:
        return board[0][0]
    
    # Check if there is no 0 in any row
    elif 0 not in board[0] and 0 not in board[1] and 0 not in board[2]:
            return 0
    else:
        return -1
```

```python
def isSolved(board):
    rows = [x for x in board]
    cols = [x for x in zip(*board)]
    diags = [r[i] for i, r in enumerate(board)]
    altdiags = [r[~i] for i, r in enumerate(board)]
    
    result = rows + cols + [diags] + [altdiags]
    
    if [1,1,1] in result:
        return(1)
    elif [2,2,2] in result:
        return(2)
    elif not any(0 in i for i in result):
        return(0)
    else:
        return(-1)
```

### [Unary function chainer](https://www.codewars.com/kata/54ca3e777120b56cb6000710/train/python)

```python
def chained(functions):
    """
    A variation of my solution that goes even further with the lambda, but with the same fundamental logic I had. 
    
    Here's the idea: You must pass a list of functions to this function, but they must then be called in order 
    on some input.
    
    To call an internal function, you can use the `function()()` notation, which returns an internal function 
    from the first call `function()` that takes the argument next to it (the leftover `()` can be `(whateverargs)`. 
    
    I accomplished this via the 'chainer` function which returns itself (making it callable) after returning a lambda 
    reduce. Here's a more verbose example:
    
    >>> def returnFunction():
    ...     def myFunction(string):
    ...         print(string)
    ...     return myFunction
    ... 
    >>> returnFunction()('test!')
    test!
    
    returnFunction just returns myFunction; it's just a wrapper function that returns a callable object that runs
    on the next argument (`string`)
    
    The next step is to run the function in order. We can do this because of a little track in the `reduce` definition 
    that lets us supply an initial value `input`. Using this trick, we can supply a whole list of functions and ask
    `reduce` to run that `initial` through every function in that list until we have a final value. 
    
    These functions can be any defined (even anonymous) function, but you must define the initial `lambda x, y: y(x)` 
    so it attempts to call function y on argument x. Here's another example:
        
    >>> reduce((lambda x, y: y(x)), [lambda x:x**2, lambda z:z*3, print], 10)
    300
    
    Here we have all kinds of functions and valeus, but they all run in order on the `initial=10` and reduce to 
    some value (which is what `reduce` does)
    
    Combining these ideas with an additional lambda in lieu of 'chainer' definition allows this line of code to run the 
    entire list of `functions` on input `x`, where `initial = x`, and functions are reduced via `f(v)` where `v` is
    the return of the previous function in the list. 
    """
    return lambda x: reduce(lambda v, f: f(v), functions, x)
```

An integral

can be approximated via Simpson's rule: 
$$\frac{b-a}{3n}(f(a)+f(b)+ 4\sum_{i=1}^{n/2}f(a+(2i-1)h)+2\sum_{i=1}^{n/2-1}f(a+2ih))$$

Here, we will translate that rule into Python code for the function $f(x) = \frac{3}{2}sin^3x$ for the interval $0, \pi$:

```python
def simpson(f, a, b, n):
    """Approximates the definite integral of f from a to b by the
    composite Simpson's rule, using n subintervals (with n even)"""

    if n % 2:
        raise ValueError("n must be even (received n=%d)" % n)

    h = (b - a) / n
    s = f(a) + f(b)

    for i in range(1, n, 2):
        s += 4 * f(a + i * h)
    for i in range(2, n-1, 2):
        s += 2 * f(a + i * h)

    return s * h / 3
```

### [Reverse or rotate?](https://www.codewars.com/kata/56b5afb4ed1f6d5fb0000991/train/python)

```python
def revrot(strng, sz):
    """
    A more elegent solution that uses a named lambda and a ternary operator to do the entire step.
    My function is still similar to the next besst practice, but is too verbose with the `parts` variable
    
    * Creates a function `func` that appends everything past index `1` to first element `0` 
    * Fun runs if the sum of the integer i for i in the list x mod 2 (my first if condition) 
    is 0, else returns the flipped version of list x
    * Ternary operator returns "" if fails any of the initial conditions, else joins a "" (like mine) for
    the output of `func()` on a list generated from what is essentially my `parts` variable
    
    A really interesting combination of lambda and ternary operators to turn this entire block
    into just two lines. I know I keep saying I should write longer code, but a lambda when used 
    correctly is just so elegant. 
    
    """
    func = lambda x : x[1:] + x[0] if sum(int(i) for i in x) % 2 else x[::-1]
    return "" if sz <= 0 or sz > len(strng) else "".join(func(strng[i:i+sz]) for i in xrange(0, len(strng) - sz + 1, sz))
```

```python
def revrot(strng, sz):
    if sz <= 0 or sz >= len(strng) or not strng:
        return ""
    else:
        final = []
        parts = [strng[i:i+sz] for i in range(0,len(strng),sz) if len(strng[i:i+sz]) == sz]
        for entry in parts:
            if sum([int(i)**3 for i in list(entry)]) % 2 == 0:
                final += entry[::-1]
            else:
                final += list(entry)[1:] + list(entry)[:1]
        return ''.join(final)
```

```python
opposite = {'NORTH': 'SOUTH', 'EAST': 'WEST', 'SOUTH': 'NORTH', 'WEST': 'EAST'}

def dirReduc(plan):
    """
    An excellent example of overcomplicating the problem. I first assumed it meant remove EVERY
    opposite direction entry in the list, which is what the function below provides. I realized
    it meant to flip every direction in order, which is where I got the second (easier) function
    after a waste of time. 
    
    This best practice function differs in the following ways:
    
    * Creation of a dictionary of flipped directions that take the input direction for easy comparison (Handled in my function as literal logic pairs)
    * Same 'if not' to avoid the empty list comparison problem (empty_list[-1] raises an Index Error)
    * Abstraction of my complex booleans to a simple dictionary flip, much more elegant
    * Same .pop and .append logic whic updates the result based on matching the first and last value
    
    Do not underestimate dictionaries; the right key-pair can avoid tons of lines of logic. This approach
    is not only just as easy/effecient to implement, but also much more logical to understand.
    
    Strive for the balance, and use dictionaries more; they're powerful. 
    """
    new_plan = []
    for d in plan:
        if new_plan and new_plan[-1] == opposite[d]:
            new_plan.pop()
        else:
            new_plan.append(d)
    return new_plan
```

### [Format a string of names like 'Bart, Lisa & Maggie'.](https://www.codewars.com/kata/53368a47e38700bd8300030d/train/python)

```python
def namelist(hashnames):
    """
    Another rare case where my approach is comprable to the best in class answer.
    
    This is a retrofit of the 'like' function defined earlier, and works only becomes of the *names parameter, which
    allows each argument to be passed in individually, instead of as a list. This is SUPER HELPFUL. 
    """
    names = [i['name'] for i in hashnames]
    return {
        0: '',
        1: '{}', 
        2: '{} & {}', 
        3: '{}, '* (len(names)-2) + '{} & {}'
    }[min(3, len(names))].format(*names)
    
```
